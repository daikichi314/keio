```python
# 1. 
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn
sns.set_theme(style="darkgrid")

# 2.  (NumPy)
# 010100
time = np.linspace(0, 10, 100)
# 
voltage = np.sin(time) + np.random.normal(0, 0.15, 100)

# 3.  (Pandas)
# 'Time''Voltage'
df = pd.DataFrame({
    'Time': time,
    'Voltage': voltage
})

# 4. 5
print("--- Data Preview ---")
print(df.head())

# 5.  (Matplotlib & Seaborn)
plt.figure(figsize=(10, 6))

# 
sns.scatterplot(data=df, x='Time', y='Voltage', color='blue', label='Measured Data')

# 
plt.plot(time, np.sin(time), color='red', linestyle='--', linewidth=2, label='Theoretical Value')

# 
plt.title('Simulation: Voltage vs Time', fontsize=15)
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend()

# 
plt.show()
```
